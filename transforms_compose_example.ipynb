```python
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
%matplotlib inline
trans_list = [transforms.CenterCrop(100), transforms.RandomHorizontalFlip(p=1)
]
my_trans = transforms.Compose(trans_list)
img = Image.open(r"F:\\img\\lena.jpg")
plt.subplot(121)
imshow(img)
my_img = my_trans(img)
plt.subplot(122)
imshow(my_img)
plt.show()
```
