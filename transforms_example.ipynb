```python
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
%matplotlib inline
trans_list = [transforms.CenterCrop(100), transforms.RandomHorizontalFlip(p=1)
]
my_trans = transforms.Compose(trans_list)
img = Image.open(r"F:\\img\\lena.jpg")
plt.subplot(121)
imshow(img)
my_img = my_trans(img)
plt.subplot(122)
imshow(my_img)
plt.show()
```

```python
img = Image.open(r"F:\\img\\cat.jpg")
plt.subplot(221)
imshow(img)
#randomly change the brightness
my_img1 = transforms.ColorJitter((0.5,0.6))(img)
plt.subplot(222)
imshow(my_img1)
#randomly change the contrast
my_img2 = transforms.ColorJitter(0,(0.5,0.6))(img)
plt.subplot(223)
imshow(my_img2)
#randomly change the saturation
my_img3 = transforms.ColorJitter(0,0,(0.5,0.6))(img)
plt.subplot(224)
imshow(my_img3)
plt.show()
```

```python
img = Image.open(r"F:\\img\\cat.jpg")
plt.subplot(231)
imshow(img)
crops = transforms.FiveCrop(100)(img)
ii = 2
for crop in crops:
    plt.subplot(2,3,ii)
    imshow(crop)
    ii += 1
```

```python
img = Image.open(r'F:\\img\\cat.jpg')
plt.subplot(1,3,1)
imshow(img)

my_img1 = transforms.Grayscale(1)(img)
plt.subplot(132)
imshow(my_img1)
print("my_img1.mode: ",my_img1.mode)

my_img2 = transforms.Grayscale(3)(img)
plt.subplot(133)
imshow(my_img2)
print("my_img2.mode: ",my_img2.mode)
```

```python
img = Image.open(r'F:\\img\\cat.jpg')
plt.subplot(121)
imshow(img)
my_img = transforms.Pad(padding=20, fill=(0,255,255), padding_mode='constant')(img)
plt.subplot(122)
imshow(my_img)
```

```python
img = Image.open(r'F:\\img\\cat.jpg')
my_img1 = transforms.RandomAffine(60)(img)
plt.subplot(221).set_title("rotate_only")
imshow(my_img1)
my_img2 = transforms.RandomAffine(60,translate=(0.3,0.3))(img)
plt.subplot(222).set_title("rotate_translate")
imshow(my_img2)
my_img3 = transforms.RandomAffine(60,scale=(2.0,2.1))(img)
plt.subplot(223).set_title("rotate_scale")
imshow(my_img3)
my_img4 = transforms.RandomAffine(0,shear=60)(img)
plt.subplot(224).set_title("shear_only")
imshow(my_img4)
plt.tight_layout()
```

```python
img = Image.open(r'F:\\img\\cat.jpg')
plt.subplot(121).set_title("origin")
imshow(img)
my_img1 = transforms.RandomPerspective()(img)
plt.subplot(122).set_title("perspective")
imshow(my_img1)
```

```python
trans_list = [transforms.ToTensor(), transforms.RandomErasing(), transforms.ToPILImage()
]
my_trans = transforms.Compose(trans_list)
img = Image.open(r"F:\\img\\lena.jpg")
plt.subplot(121)
imshow(img)
my_img = my_trans(img)
plt.subplot(122)
imshow(my_img)
plt.show()
```

```python
import numpy as np
img = Image.open(r"F:\\img\\lena.jpg")
plt.subplot(121)
imshow(img)

def my_crop(img,crop_size=100):
    img = np.asarray(img,dtype='uint8')
    img = img[:99,:99]
    return Image.fromarray(img.astype('uint8'))

my_img = transforms.Lambda(lambda img: my_crop(img, crop_size=100))(img)
plt.subplot(122)
imshow(my_img)
plt.show()
```
